```python
import random
import ast
import pandas as pd
import numpy as np


def calc_sums(row):
    data = []
    try:
        lst = list(row['prec12'])
        
        for j in range(12):
            d = np.asarray(lst[j*12:j*12+12])
            d = np.where(d == 65535, 0, d)
            d = np.where(d == np.nan, 0, d)
            d = np.where(d < 0, 0, d)
            data.append(np.sum(d))
            
    except:
        data = [0]*12
        
    return data

def avg_height(row):
    h = row['height_dtm']
    h = np.where(h > 1000, np.nan, h)
    if len(h) == 0:
        return np.nan
    return np.nanmean(h)

def avg_height2(row):
    h = row['height_dtm'][10:31,10:31]
    h = np.where(h > 1000, np.nan, h)
    if len(h) == 0:
        return np.nan
    return np.nanmean(h)

def avg_height3(row):
    h = row['height_dtm'][15:26,15:26]
    h = np.where(h > 1000, np.nan, h)
    if len(h) == 0:
        return np.nan
    result = np.nanmean(h)
    if np.isnan(result):
        return 200
    else:
        return result
    
def avg_height4(row):
    h = row['height_dtm'][18:25,18:25]
    h = np.where(h > 1000, np.nan, h)
    if len(h) == 0:
        return np.nan
    result = np.nanmean(h)
    if np.isnan(result):
        return 200
    else:
        return result

def avg_height5(row):
    h = row['height_dtm'][21:22,21:22]
    h = np.where(h > 1000, np.nan, h)
    if len(h) == 0:
        return np.nan
    result = np.nanmean(h)
    if np.isnan(result):
        return 200
    else:
        return result

def count(row):
    h = row['height']
    print(h.shape)
    return np.count_nonzero(h < 1000)

df = pd.read_pickle('test_samples_close_homes250_500.pkl')

df = df.dropna(axis=0)
# df
df = df.drop(['index'], axis=1)
df['prec_sums'] = df.apply(calc_sums, axis=1)
cols = ['sum'+str(i+1) for i in range(12)]
df[cols] = pd.DataFrame(df.prec_sums.tolist(), index=df.index)
df['past3hours'] = df['sum1'] + df['sum2'] + df['sum3']
df = df[(df.past3hours > 1500)]

df['avg_height'] = df.apply(avg_height, axis=1)
df['avg_height2'] = df.apply(avg_height2, axis=1)
df['avg_height3'] = df.apply(avg_height3, axis=1)
df['avg_height4'] = df.apply(avg_height4, axis=1)
df['avg_height5'] = df.apply(avg_height5, axis=1)
# df['count'] = df.apply(count, axis=1)
df
```

```python
# df1 = (df[df.target==1])
# df0 = (df[df.target==0])
# # df0 = df0.sample(len(df1))
# df = pd.concat([df0, df1])
# # values = list(df[df.target==0]['past3hours'])
# # len(values)

def get_flow(data):
    dtm = data['height_dtm']
    dsm = data['height_dsm']
    
    #water
    dtm[dsm>1000] = -100
    
    amount_of_water = dtm.shape[0] * dtm.shape[1]
    building_count = np.sum(dtm>1000)
    amount_of_rain_per_cell = amount_of_water/(amount_of_water-building_count)/1000
    
    # simple alg
    result = np.full(dtm.shape, 0.0)
    for i in range(60):
        if i < 20:
            result[dtm<1000] += amount_of_rain_per_cell
            
        for x in range(1, dtm.shape[1]-1):
            for y in range(1, dtm.shape[0]-1):
                if dtm[y,x] < 1000:
                    val = dtm[y,x]
                    waterval = result[y,x]


                    topdiff = val+waterval-dtm[y+1,x]+result[y+1,x]
                    bottomdiff = val+waterval-dtm[y-1,x]+result[y-1,x]
                    leftdiff = val+waterval-dtm[y,x-1]+result[y,x-1]
                    rightdiff = val+waterval-dtm[y,x+1]+result[y,x+1]
                    water_to_move = min(waterval, max([topdiff, bottomdiff, leftdiff, rightdiff]))
                    total = 0
                    if topdiff > 0:
                        total+=topdiff

                    if leftdiff > 0:
                        total+=leftdiff

                    if rightdiff > 0:
                        total+=rightdiff

                    if bottomdiff > 0:
                        total+=bottomdiff

                    if topdiff > 0:
                        result[y+1,x] += (topdiff/total) * water_to_move

                    if leftdiff > 0:
                        result[y,x-1] += (leftdiff/total) * water_to_move

                    if rightdiff > 0:
                        result[y,x+1] += (rightdiff/total) * water_to_move

                    if bottomdiff > 0:
                        result[y-1,x] += (bottomdiff/total) * water_to_move

                    if total > 0:
                        result[y,x] -= water_to_move
                    
    result[dsm>1000] = 0
    return result


df['rain_flow'] = df.apply(get_flow, axis=1)

```
